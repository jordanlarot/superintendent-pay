```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://transparentcalifornia.com/salaries/search/?q=superintendent&y=2020&page=1'
r = requests.get(url)
html_doc = r.text
soup = BeautifulSoup(html_doc, 'lxml')
```

```python
# Get table head
tbody = soup.tbody

print(tbody.prettify())
```

```python
trs = tbody.contents
```

```python
tds = trs.find_all('td')
```

```python
for tr in trs:
    print(tr)
```
