```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://keithgalli.github.io/web-scraping/webpage.html'
r = requests.get(url)
html_doc = r.text
soup = BeautifulSoup(html_doc, 'lxml')
```

```python
links = soup.find_all('a')

for link in links:
    print(link['href'])
```

```python
https = soup.select('a[href^="https"]')
htmls = soup.select('a[href$=".html"]')


#links = list(https).append(list(htmls))

for http in https:
    print(http['href'])
    
for html in htmls:
    print(html['href'])
```

```python
import re

yt = soup.find_all('a', string=re.compile("k"))

for i in yt:
    print(i['href'])
```

```python
social = soup.find('ul', class_='socials')
social = social.find_all('a')

for i in social:
    print(i['href'])
```
